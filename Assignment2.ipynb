# Univariate Analysis

# Pdf

```python
features = [i for i in data.columns]
del features[0]
```

```python
label =data['Date']
for i in features:
  sns.FacetGrid(data,size=4).map(sns.distplot,f"{i}").add_legend()
  plt.title(f"{i} vs Density")
  plt.show()
```

1.   As we can see all the cases rised and increased 
2.   As time elapsed the death and recovery rate also 
increased
3.   We can see that the number of the new cases have a bump means there was a sharp rise at that 
4.  Similarly the case of the new deaths had increase all of sudden then its density fell.
5.  As we can see that deaths/100 increased and was having a level for sometime
6. Also the recovery rate declined for a while probably at the time of increase of the death rate incerment
7.  IN the very end we can see that the number of countries that got affected by it got decreased but then it got a hike and then it started falling.


# CDF

```python
for i in features:
  count,binEdges=np.histogram(data[i],bins=10,density=True)
  pdf=count/sum(count)
  print(pdf,"\n",binEdges)
  cdf = np.cumsum(pdf)
  plt.plot(binEdges[1:],cdf,label='cdf')
  plt.xlabel(f"{i}")
  plt.title(f"CDF PLOT FOR THE {i}")
  plt.legend()
  plt.show()
```

1. As we can see that the number of cases,recovered number of the people and the death rate,active cases grown in almost linear fashion.

2. Coming to the new cases it we can see it was growing slowly but got a sudden hike and so the new deaths.
3. As we can see the Deaths/100 Recovered also got a hike.
4. In the very end we can see that the number of countries that were getting affected were slow earlier but later on it got pace.
